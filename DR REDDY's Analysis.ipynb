```python
#EMA
dff['EMA_9'] = dff['Close'].ewm(9).mean().shift()
dff['EMA_22'] = dff['Close'].ewm(22).mean().shift()
#SMA
dff['SMA_5'] = dff['Close'].rolling(5).mean().shift()
dff['SMA_10'] = dff['Close'].rolling(10).mean().shift()
dff['SMA_15'] = dff['Close'].rolling(15).mean().shift()
dff['SMA_30'] = dff['Close'].rolling(30).mean().shift()
#RSI14
def RSI(dff, n=14):
    close = dff['Close']
    delta = close.diff()
    delta = delta[1:]
    pricesUp = delta.copy()
    pricesDown = delta.copy()
    pricesUp[pricesUp < 0] = 0
    pricesDown[pricesDown > 0] = 0
    rollUp = pricesUp.rolling(n).mean()
    rollDown = pricesDown.abs().rolling(n).mean()
    rs = rollUp / rollDown
    rsi = 100.0 - (100.0 / (1.0 + rs))
    return rsi

dff['RSI'] = RSI(dff).fillna(0)

EMA_12 = pd.Series(dff['Close'].ewm(span=12, min_periods=12).mean())
EMA_26 = pd.Series(dff['Close'].ewm(span=26, min_periods=26).mean())
dff['MACD'] = pd.Series(EMA_12 - EMA_26)
dff['MACD_signal'] = pd.Series(dff.MACD.ewm(span=9, min_periods=9).mean())
```

  - <p style = "color:#226666">Moving average is one of the basic concepts in technical analysis and is widely used in financial markets. This concept helps in analyzing price patterns and trends. Moving averages are usually divided into two types: simple moving average (SMA) and exponential moving average (EMA).
Simple Moving Average (SMA):
In SMA, to calculate the average, the prices are considered equally and without weight priority.
To calculate the SMA with a time window n, we add the prices of the previous n periods and divide by their number.
Exponential Moving Average (EMA):
In the EMA, new prices are given more weight. This means that more recent prices are given more weight in the calculation of the average.
A regression relation is used to calculate the EMA with an n time window.
A moving average is used as an indicator of changes in price trends. In technical analysis, the comparison of moving averages with the current price or with each other can be interpreted as buy or sell signals. For example, if the price crosses above the SMA, this may be interpreted as a buy signal.
If the SMA (Simple Moving Average) chart is moving up, it can mean several things and depends on the market context and time period under consideration. In technical analysis, the upward movement of the SMA is one of the indicators of the strength of the market's upward trend. In this scenario, you can consider it as a positive signal.
Here are some possible interpretations for the upward movement of the SMA:
The strength of the upward trend:
If the SMA moves up, it is an indication of the strength of the market's uptrend. This may indicate that the average prices have increased over a certain time period (eg 5, 10, 50 days) and that prices are generally moving in an upward direction.
Motion stability:
An upward movement of the SMA may indicate trend persistence, meaning that prices are continuing to rise and this movement is likely to continue for some time.
Buy signal:
Some traders interpret the upward SMA as a buy signal. This may mean that the market is strengthening and prices may move positively.
Now, you should note that any analysis should be done carefully with additional information and decisions should be made based on market conditions and existing risks. Also, using other indicators and analytical tools along with SMA can improve decision making.
Examining the simple moving average (SMA) in different time frames plays an important role in the technical analysis of financial markets. Using SMA with different timeframes means the following:
9-day SMA:
SMA with this timeframe usually represents a short-term trend and is more sensitive to price changes. As a result, this SMA performs best in detecting rapid price changes and identifying entry and exit points for short-term investment positions.
22-day SMA:
SMA with this timeframe is suitable for displaying medium-term trends and detecting medium-term price changes. This may be useful for investors looking for medium-term entry and exit analysis.
By using SMA with different timeframes, investors can identify strengths and weaknesses in price trends and make decisions based on technical analysis tailored to your own investment goals and style. Also, using multiple SMAs with different timeframes as "twin peaks" is also used to confirm signals
If the Simple Moving Average (SMA) and the Exponential Moving Average (EMA) are close together and similar on a chart, this may indicate a convergence in the price trend. This situation may mean confirmation of the market trend.
Some traders and analysts use congruence between moving averages as a sign of trend strength and continuity. This means that if both types of moving averages move close to each other, it confirms that the price trend is ongoing and short-term changes may be small and If the Simple Moving Average (SMA) and the Exponential Moving Average (EMA) both appear to be moving close to each other and moving in the same direction (up or down) at the same time, it can indicate strength and confirmation of the market trend. ```python

```python
# Calculate the 20-day moving average
dff['MA_20'] = dff['Adj Close'].rolling(window=20).mean()

# Calculate the 20-day standard deviation
dff['Upper_Band'] = dff['MA_20'] + 2 * dff['Adj Close'].rolling(window=20).std()
dff['Lower_Band'] = dff['MA_20'] - 2 * dff['Adj Close'].rolling(window=20).std()


fig, (ax1, ax2,ax3) = plt.subplots(3, 1, sharex=True, figsize=(10, 8))


ax1.plot(dff['Date'], dff['Adj Close'], label='Stock Price', marker='o')
ax1.set_ylabel('Adj Close')
ax1.legend()
ax2.plot(dff['Date'], dff['MA_20'], label='20-Day Moving Average', linestyle='--', marker='o', color='green')
ax2.set_ylabel('Adj Close')
ax2.legend()


ax3.plot(dff['Date'], dff['Upper_Band'], label='Upper Bollinger Band', linestyle='--', color='green')
ax3.plot(dff['Date'], dff['Lower_Band'], label='Lower Bollinger Band', linestyle='--', color='red')
ax3.set_xlabel('Date')
ax3.set_ylabel('Bollinger Bands')
ax3.legend()

plt.suptitle('Bollinger Bands Chart with Subplots')
plt.show()
```

When the price approaches or crosses the upper band, the market may be recognized as "overbought" and the price is likely to decline in the future. Similarly, when the price approaches or crosses the lower band, the market may be recognized as "oversold" and the price is likely to rise in the future.</p>
